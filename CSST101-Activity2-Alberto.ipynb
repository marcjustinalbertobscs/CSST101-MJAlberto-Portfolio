### Problem 1: Socrates

**Statements to translate:**
1.  'All humans are mortal.'
2.  'Socrates is a human.'

**Statement to prove:**
3.  'Socrates is mortal.'

**1. Define Predicates and Constants:**
*   `Human(x)`: x is a human
*   `Mortal(x)`: x is mortal
*   `Socrates`: a constant representing Socrates

**2. Translate Statements into Predicate Logic:**
*   **Premise 1 (P1):** `∀x (Human(x) → Mortal(x))`
*   **Premise 2 (P2):** `Human(Socrates)`
*   **Conclusion (C):** `Mortal(Socrates)`

**3. Proof using Natural Deduction:**

| Step | Formula                                 | Derivation           |
| :--- | :-------------------------------------- | :------------------- |
| 1    | `∀x (Human(x) → Mortal(x))`             | P1                   |
| 2    | `Human(Socrates) → Mortal(Socrates)`    | Universal Instantiation (UI) from 1 |
| 3    | `Human(Socrates)`                       | P2                   |
| 4    | `Mortal(Socrates)`                      | Modus Ponens (MP) from 2, 3 |

**The conclusion 'Socrates is mortal.' is successfully proven.**

### Problem 2: Juan

**Statements to translate:**
1.  'Every student who studies passes the exam.'
2.  'Juan is a student and he studies.'

**Statement to prove:**
3.  'Juan passes the exam.'

**1. Define Predicates and Constants:**
*   `Student(x)`: x is a student
*   `Studies(x)`: x studies
*   `PassesExam(x)`: x passes the exam
*   `Juan`: a constant representing Juan

**2. Translate Statements into Predicate Logic:**
*   **Premise 1 (P1):** `∀x ((Student(x) ∧ Studies(x)) → PassesExam(x))`
*   **Premise 2 (P2):** `Student(Juan) ∧ Studies(Juan)`
*   **Conclusion (C):** `PassesExam(Juan)`

**3. Proof using Natural Deduction:**

| Step | Formula                                             | Derivation           |
| :--- | :-------------------------------------------------- | :------------------- |
| 1    | `∀x ((Student(x) ∧ Studies(x)) → PassesExam(x))`   | P1                   |
| 2    | `(Student(Juan) ∧ Studies(Juan)) → PassesExam(Juan)`| Universal Instantiation (UI) from 1 |
| 3    | `Student(Juan) ∧ Studies(Juan)`                    | P2                   |
| 4    | `PassesExam(Juan)`                                  | Modus Ponens (MP) from 2, 3 |

**The conclusion 'Juan passes the exam.' is successfully proven.**

### Problem 3: Mark

| 2    | `Teacher(Mark) → ∃y (Student(y) ∧ Advises(Mark, y))`          | Universal Instantiation (UI) from 1 |
| 3    | `Teacher(Mark)`                                                 | P2                   |
| 4    | `∃y (Student(y) ∧ Advises(Mark, y))`                            | Modus Ponens (MP) from 2, 3 |

**The conclusion 'Mark advises at least one student.' is successfully proven.**

```
from sympy.logic.api import And, Implies, ForAll
from sympy.abc import x
from sympy.logic.predicates import Predicate, Function

# 1. Define Predicates and Constants
Human = Predicate('Human')
Mortal = Predicate('Mortal')
Socrates = Function('Socrates')

# 2. Translate Statements into Predicate Logic
# Premise 1 (P1): ∀x (Human(x) → Mortal(x))
P1 = ForAll(x, Implies(Human(x), Mortal(x)))

# Premise 2 (P2): Human(Socrates)
P2 = Human(Socrates)

# Conclusion (C): Mortal(Socrates)
C = Mortal(Socrates)

print("Premise 1:", P1)
print("Premise 2:", P2)
print("Conclusion:", C)

# To demonstrate Universal Instantiation (UI) from P1 for Socrates:
# We can substitute 'Socrates' for 'x' in the universal statement.
# This step is often implicit in manual proofs but can be represented symbolically.
universal_instantiation_step = Implies(Human(Socrates), Mortal(Socrates))
print("\nStep 2 (UI from P1):", universal_instantiation_step)

# To demonstrate Modus Ponens (MP) from P2 and the UI step:
# If (Human(Socrates) -> Mortal(Socrates)) and Human(Socrates) are true,
# then Mortal(Socrates) must be true.
print("\nStep 4 (MP from UI step and P2): Resulting conclusion is", C)
```
